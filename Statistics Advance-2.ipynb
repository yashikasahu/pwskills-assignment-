### **Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)?**  

🔹 **PMF (Probability Mass Function)** – Used for **discrete random variables**. It gives the probability of each specific value.  

📌 **Example (Rolling a Fair Die):**  
\[
P(X = x) = \frac{1}{6}, \quad x \in \{1,2,3,4,5,6\}
\]  

🔹 **PDF (Probability Density Function)** – Used for **continuous random variables**. It represents the probability density at a given value but not the probability itself.  

📌 **Example (Normal Distribution PDF):**  
\[
f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x - \mu)^2}{2\sigma^2}}
\]  
where \( \mu \) is the mean and \( \sigma \) is the standard deviation.  

📌 **Key Difference:** PMF gives probabilities for **specific values**, while PDF gives a **density** over an interval.

### **Q2: What is the Cumulative Density Function (CDF)? Why is it Used?**  

🔹 **CDF** gives the probability that a random variable \( X \) takes a value **less than or equal** to a given number.  

📌 **Formula:**  
\[
F(x) = P(X \leq x)
\]  

📌 **Example (Normal Distribution CDF):**  
For a **standard normal distribution** (\( \mu = 0, \sigma = 1 \)):  
\[
P(X \leq 1) = 0.8413
\]  

🔹 **Why is CDF Used?**  
✔ Helps compute probabilities over **ranges**.  
✔ Used for **percentiles & quantiles**.  
✔ Provides a **cumulative** probability measure.

### **Q3: Situations Where Normal Distribution is Used & Its Parameters**  

📌 **Examples of Normal Distribution Usage:**  
1️⃣ Heights of people  
2️⃣ Blood pressure readings  
3️⃣ Test scores (SAT, IQ)  
4️⃣ Measurement errors  

📌 **Effect of Parameters on Shape:**  
- **Mean \( \mu \)** – Determines the **center** of the distribution.  
- **Standard Deviation \( \sigma \)** – Controls the **spread**.  
   - Large \( \sigma \) → Wide curve  
   - Small \( \sigma \) → Narrow curve  

📌 **Key Property:**  
- **68-95-99.7 Rule** → **68% of data** lies within \( 1\sigma \), **95% within 2\sigma**, and **99.7% within 3\sigma**.

### **Q5: What is Bernoulli Distribution? How is it Different from Binomial?**  

🔹 **Bernoulli Distribution:**  
- A **discrete** distribution with **only two possible outcomes**: Success (\(1\)) or Failure (\(0\)).  
- Probability of success: \( p \), Probability of failure: \( 1 - p \).  

📌 **Example:**  
- Flipping a coin **once**:  
  \[
  P(X = 1) = p, \quad P(X = 0) = 1 - p
  \]  

🔹 **Difference Between Bernoulli & Binomial:**  

| Feature         | Bernoulli Distribution | Binomial Distribution |
|----------------|-----------------------|-----------------------|
| **Number of Trials** | 1 | \( n \) |
| **Possible Outcomes** | 0 or 1 | 0 to \( n \) |
| **Example** | A single coin flip | 10 coin flips |

### **Q6: Probability of Selecting a Value Greater than 60 in Normal Distribution**  

📌 **Given:**  
- \( \mu = 50 \), \( \sigma = 10 \)  
- Find \( P(X > 60) \).  

📌 **Step 1: Convert to Z-score**  
\[
Z = \frac{X - \mu}{\sigma} = \frac{60 - 50}{10} = 1
\]  

📌 **Step 2: Find \( P(Z > 1) \) from Normal Table**  
\[
P(Z \leq 1) = 0.8413
\]  
\[
P(Z > 1) = 1 - 0.8413 = 0.1587
\]  

✅ **Answer:** **15.87% probability**.

### **Q7: What is Uniform Distribution?**  

🔹 **Uniform Distribution** – A distribution where **all values have equal probability**.  

📌 **Example:** Rolling a **fair die**  
\[
P(X = x) = \frac{1}{6}, \quad x \in \{1,2,3,4,5,6\}
\]  

📌 **Continuous Uniform Example:**  
- Generating a **random number** between **0 and 1**.  
- Every number in **[0,1]** is equally likely.

### **Q8: What is the Z-score? Why is it Important?**  

📌 **Definition:**  
The **Z-score** tells how many standard deviations a value is **from the mean**.  

📌 **Formula:**  
\[
Z = \frac{X - \mu}{\sigma}
\]  

📌 **Importance:**  
✔ Standardizes different datasets for **comparison**.  
✔ Helps in **outlier detection**.  
✔ Used in **normal probability calculations**.  

📌 **Example:** If \( Z = 2 \), the value is **2 standard deviations** above the mean.

### **Q9: What is the Central Limit Theorem (CLT)? Why is it Important?**  

📌 **Definition:**  
The **Central Limit Theorem (CLT)** states that the **sampling distribution of the mean** will approach a **normal distribution** as the **sample size increases**, regardless of the population’s original distribution.  

📌 **Importance:**  
✔ Allows **parametric statistical tests** (e.g., t-tests, confidence intervals).  
✔ Works even if the **original population is not normal**.  
✔ Justifies using the **Normal Distribution** in real-world problems.

📌 **Example:**  
- Taking samples from a **skewed** population → The **mean of samples** follows a **normal distribution**.

---

### **Q10: Assumptions of the Central Limit Theorem**  

🔹 **1. Random Sampling**  
- The samples must be **randomly chosen** from the population.  

🔹 **2. Independent Observations**  
- Each observation **should not** influence another.  

🔹 **3. Sample Size is Large**  
- Typically, **\( n > 30 \)** is considered **sufficient** for CLT to hold.  

🔹 **4. Finite Variance**  
- The population **must have a finite variance** (not infinite).
