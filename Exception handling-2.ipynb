Q2. Write a python program to print Python Exception Hierarchy.

```python
def exception_hierarchy(exception_class, indent=0):
    print(' ' * indent + exception_class.__name__)
    for subclass in exception_class.__subclasses__():
        exception_hierarchy(subclass, indent + 4)
exception_hierarchy(BaseException)
```

The ArithmeticError class in Python is a base class for all exceptions that are related to arithmetic operations. It is a subclass of the

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

The LookupError class is not meant to be raised directly; instead, it is the superclass for more specific 
lookup-related exceptions like IndexError and KeyError. The main purpose of LookupError is to allow for a 
single exception handler to catch multiple lookup-related errors at once.

```python
# KeyError:Raised when trying to access a dictionary key that does not exist.
my_dict = {'name': 'Alice'}
print(my_dict['age'])
```

```python
#IndexError:Raised when trying to access an element from a list or tuple using an index that is out of range.
my_list = [1, 2, 3]
print(my_list[5])
```

Q5. Explain ImportError. What is ModuleNotFoundError?

ImportError is a built-in exception in Python that is raised when an import statement fails to successfully 
import a module or a specific attribute from a module

ModuleNotFoundError is a specific exception in Python that is raised when an attempt to import a module fails because the module cannot be found.
